## 데이터셋 가공

```python
data_set = data_set.drop(['Unnamed: 0', '계약일'], axis = 1)
```

```python
df = data_set.drop(data_set[data_set['도로명'] == ' '].index)
```

```python
df
```

```python
doro = []
doro_gil = []

delete_idx = []

for i, dong_list in enumerate(list(df['도로명'].str.split())):
  if len(dong_list) == 2:
    doro.append(dong_list[0])
    doro_gil.append(dong_list[1])
  else:
    doro.append('')
    doro_gil.append(dong_list[0])

print(len(doro))
print(len(doro_gil))

print(delete_idx)
```

```python
df['도로명'] = doro
df['건물번호'] = doro_gil
```

```python
df = df.drop(df[df['번지'].str.contains('가')].index)
```

```python
df
```

```python
price = []


for i, p_list in enumerate(list(df['거래금액(만원)'].str.split(','))):
  if len(p_list) == 3:
    price.append(int(p_list[0]) * 1000000 + int(p_list[1]) * 1000 + int(p_list[2]))
  elif len(p_list) == 2:
    price.append(int(p_list[0]) * 1000 + int(p_list[1]))
  elif len(p_list) == 1:
    price.append(int(p_list[0]))
  else:
    print(p_list)
```

```python
df['거래금액(만원)'] = price
```

```python
len(price)
```

```python
df = df.drop(['번지'], axis = 1)
```

```python
df.to_csv('data/dataset_final.csv')
```
